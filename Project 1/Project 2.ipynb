```python
# Commented out IPython magic to ensure Python compatibility.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# %matplotlib inline
```

```python

data['Churn'].value_counts().head().plot.bar()

sizes=data['Churn'].value_counts(sort=True)
labels=np.unique(data.Churn)
plt.figure(figsize=(6,6))
plt.subplot(212)
plt.title("Customer churn rate")
plt.pie(sizes, labels=labels, autopct='%1.1f%%')

sns.boxplot(data,x="Churn", y="TotalCharges")
plt.show()
```

```python
# Generate a histogram to visualize the distribution of a specific column
plt.figure(figsize=(10, 6))
data['TotalCharges'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Distribution of Total Charges")
plt.xlabel("Total Charges")
plt.ylabel("Frequency")
plt.grid(False)
plt.show()
```

```python
# Generate a scatter plot to visualize the relationship between two columns
plt.figure(figsize=(10, 6))
plt.scatter(data['MonthlyCharges'], data['TotalCharges'], alpha=0.6, color='blue', edgecolor='k')
plt.title("Scatter Plot of Monthly Charges vs Total Charges")
plt.xlabel("Monthly Charges")
plt.ylabel("Total Charges")
plt.grid(True)
plt.show()
```

```python
# Convert 'SeniorCitizen' to numeric if needed (already 0/1, so okay)
# Select only numerical columns for correlation
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
plt.figure(figsize=(10, 6))
sns.heatmap(data[numeric_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
categorical_cols = ['gender', 'SeniorCitizen', 'Partner', 'Dependents',
                    'PhoneService', 'MultipleLines', 'InternetService',
                    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
                    'TechSupport', 'StreamingTV', 'StreamingMovies',
                    'Contract', 'PaperlessBilling', 'PaymentMethod', 'Churn']

for col in categorical_cols:
    plt.figure(figsize=(6, 4))
    sns.countplot(data=data, x=col, palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
numeric_cols = ['tenure', 'MonthlyCharges', 'TotalCharges']

for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(data[col], kde=True, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()
    plt.show()
```

```python
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=data, x='Churn', y=col, palette='pastel')
    plt.title(f'{col} by Churn')
    plt.tight_layout()
    plt.show()
```

```python
# Convert 'TotalCharges' to numeric and drop NaNs
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
data.dropna(inplace=True)

# Select only numeric columns
numeric_df = data.select_dtypes(include=['float64', 'int64'])
sns.clustermap(numeric_df.corr(), cmap='coolwarm', annot=True, figsize=(10, 8))
plt.title('Clustermap of Feature Correlations', y=1.05)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

# Standardize the numeric data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(numeric_df)

# Create a dataframe with scaled data
scaled_df = pd.DataFrame(scaled_data, columns=numeric_df.columns)

# Create a clustermap of customers
sns.clustermap(scaled_df, metric='euclidean', method='ward',
               cmap='viridis', figsize=(12, 10))
plt.title('Customer Clustermap (based on numerical features)', y=1.05)
plt.show()
```

```python
# Select only numeric columns + Churn
pairplot_df = data[['tenure', 'MonthlyCharges', 'TotalCharges', 'Churn']]

# Plot pairplot
sns.pairplot(pairplot_df, hue='Churn', palette='Set1', diag_kind='kde')
plt.suptitle('Pair Plot of Numerical Features by Churn', y=1.02)
plt.show()
```
